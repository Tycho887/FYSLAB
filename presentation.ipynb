```python
import termopy as tp

O2 = tp.Fluid(P=1e5, V=1e-3, T=500, gas='O2')

O2.isothermal(P=2e5)
O2.adiabatic(T=600)
O2.isobaric(V=1e-3)

tp.plot(O2, display="PV")
tp.show()

```

```python
cycle1 = tp.Carnot(T_hot=400,T_cold=300,compression_ratio=2,P=1e5,V=1e-3,gas='Octane')

print(cycle1.efficiency)

tp.plot(cycle1,display="VP")
tp.show()
```

```python
stirling = tp.Stirling(T_hot=1000,T_cold=300,compression_ratio=20,P=1e5,V=1e-3,gas='CO2')

print(stirling.efficiency)

tp.plot(stirling,display="VP")
tp.show()

```
