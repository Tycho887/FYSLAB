```python
import termoPy as tp

stirling_air = tp.Stirling(T_cold=300, T_hot=400, compression_ratio=5,V=1e-3, P=1e6, gas='Air')
stirling_argon = tp.Stirling(T_cold=300, T_hot=400, compression_ratio=5,V=1e-3, P=1e6, gas='Argon')
```

```python
tp.plot(stirling_air, display="PV")
tp.show()
```

```python
tp.plot(stirling_argon, display="PV")
tp.show()
```

## Difference in efficiency between the two solutions

```python
print("The efficiency of the Stirling engine with air is: ", stirling_air.efficiency)
print("The efficiency of the Stirling engine with argon is: ", stirling_argon.efficiency)
```
